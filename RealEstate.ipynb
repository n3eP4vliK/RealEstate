## Data Preprocessing: ##

```python
#to make the data more comfortable to work with,
# we will display the columns and the percentage of missing values in each of them.
pd.DataFrame(round(data.isna().mean()*100,)).style.background_gradient('coolwarm')
```

```python
#Let's examine the presence of explicit duplicates.
data.duplicated().sum()
```

```python
#To make it visually pleasing, let's change the name of the column 'cityCenters_nearest.
data.rename(columns={'cityCenters_nearest': 'city_centers_nearest'}, inplace=True)
```

```python
# Change the data type to datetime in the 'first_day_exposition' column.
# Remove the time component as it is consistently '00:00:00' in all rows.
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d')
data['first_day_exposition']
```

```python
# Fill missing values where possible.
# If the seller did not specify the number of balconies or nearby parks/ponds, 
# it's likely that there are none. So, we replace such missing values with 0.
# If it's not possible to propose a logical replacement, we leave the values empty.

for column in ['balcony', 'parks_around3000', 'ponds_around3000']:
    data[column] = data[column].fillna(0)
```

```python
#Let's check the result.

data.isna().sum()
```

```python
# Replace missing values in the 'is_apartment' column with booleans.
data['is_apartment'] = data['is_apartment'].fillna(False)
```

```python
# Similar action will be taken for missing values in 'floors_total' and 'locality_name'.
data['locality_name'].dropna(inplace=True)
data['floors_total'].dropna(inplace=True)
```

```python
# Round the 'days_exposition' column.
data['days_exposition'] = data['days_exposition'].round()

# Change the data type in selected columns.
for column in ['floors_total', 'balcony', 'parks_around3000', 'ponds_around3000', 'days_exposition']:
    data[column] = data[column].astype('Int64')
```

```python
# Let's make sure the changes were successful.
data.info()
display(data.head(10))
```

```python
# Sort and display unique values in the 'locality_name' column.
data['locality_name'].sort_values().unique()
```

```python
# Check if there are duplicates in the 'locality_name' column due to differences in letter case.
locality_name_lower = data['locality_name'].str.lower()
if len(locality_name_lower.unique()) == len(data['locality_name'].unique()):
    print('The number of unique values did not change when converted to lowercase.')
```

```python
# Replace names in the 'locality_name' column. Replace 'ё' with 'е' and 'городской поселок' with 'поселок городского типа'.
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
data['locality_name'] = data['locality_name'].str.replace('городской поселок', 'поселок городского типа')
data['locality_name'].sort_values().unique()
```

As we have previously established, there are 49 missing values in the 'locality_name' column, which represents the names of populated areas. It is necessary to delete the rows with missing values in this column because without the name of the populated area, all other data becomes meaningless

```python
data = data.dropna(subset=['locality_name']).reset_index(drop=True)
data.isna().sum()
```

## Calculations and Data Enhancement: ##

```python
# Add new columns to the table.

# Price per square meter:
data['one_square_meter_price'] = data['last_price'] / data['total_area']

# Day of the week when the ad was published (0 — Monday, 1 — Tuesday, and so on):
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday

# Month of the ad publication:
data['month_exposition'] = data['first_day_exposition'].dt.month

# Year of the ad publication:
data['year_exposition'] = data['first_day_exposition'].dt.year

# Type of floor in the apartment (values — 'first,' 'last,' 'other'):
def typefloor(data):
    try:
        if data['floor'] == 1:
            return 'first'
        elif data['floor'] == data['floors_total']:
            return 'last'
        else:
            return 'other'
    except:
        return 'no data'
data['type_of_floor'] = data.apply(typefloor, axis=1)

# Distance to the city center in kilometers (convert from meters to kilometers and round to whole values):
data['city_centers_nearest_km'] = (data['city_centers_nearest'] / 1000).round()


data.head(10)
```

```python
# Write a function that categorizes long sales in the 'days_exposition' column into categories:
# from 1 day to 45 days - fast sales
# 45 days and more - long sales

def categorize_days_exposition_long(days):
    try:
        if 1.0 <= days < 45.0:
             return 'fast sale'
        elif days >= 45.0:
            return 'long sale'
    except:
        pass

# Create a column with these categories and display the count of values for each of them:
data['days_exposition_length_category'] = data['days_exposition'].apply(categorize_days_exposition_long)
data['days_exposition_length_category'].value_counts()

# Plot a histogram of the category counts with a specified figure size and bin count.
data['days_exposition_length_category'].hist(figsize=(12, 5), bins=50)
plt.show()

# The longer the duration, the lower the count of values.
```
